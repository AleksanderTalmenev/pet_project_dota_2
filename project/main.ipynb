```python
import pandas as pd
import numpy as np

import seaborn as sns
import requests
```

```python
tournament_id = 6880562712
count_match = 10

list_matches = []

for id in range(tournament_id, tournament_id + count_match):
    print(id)
    
    r = requests.get(f'https://api.opendota.com/api/matches/{id}')
    dict_data_match = r.json()
    del dict_data_match['chat']
    #print(dict_data_match)
    if dict_data_match.get('error') == 'Not Found':
        continue
    df_data = pd.DataFrame([dict_data_match])
    dict_data = dict()

    for i in range(10):
        player = df_data['players'][i]
        team = 'Radiant' if player['isRadiant'] else 'Dire'
        for element in player:
            dict_data[f'pl_{i}_{team}_{element}'] = player[element]

    dict_data_match = r.json()
    del dict_data_match['players'] 
    dict_data = dict_data_match | dict_data
    list_matches.append(dict_data)

data = pd.DataFrame(list_matches)
```
